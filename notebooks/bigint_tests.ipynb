# TESTING BITWISE XOR OF LARGE PYTHON INTEGERS

Generate two random bitfields and with their corresponding integers.

```python
import src.input.bitfields as bit

size = 10000

rand_bitfield_1 = bit.rand_bitfield(size)
rand_bitfield_2 = bit.rand_bitfield(size)

rand_bigint_1 = bit.from_bitfield(rand_bitfield_1)
rand_bigint_2 = bit.from_bitfield(rand_bitfield_2)
```

Perform a XOR for the bitfields and integers.

```python
xorred_bitfield = bit.xor_bitfields(rand_bitfield_1, rand_bitfield_2)

print(rand_bitfield_1)
print(rand_bitfield_2)
print(xorred_bitfield)

xorred_bigint = rand_bigint_1 ^ rand_bigint_2

print(rand_bigint_1)
print(rand_bigint_2)
print(xorred_bigint)
```

Convert the resulting integer to a bitfield and compare it with the resulting bitfield.

```python
control_bitfield = bit.to_bitfield(xorred_bigint, size)
print(bit.compare_bitfields(control_bitfield, xorred_bitfield))
```

Convert the resulting bitfield to an integer and compare it with the resulting integer.

```python
control_number = bit.from_bitfield(xorred_bitfield)
print(control_number == xorred_bigint)

```
