```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
# model function
def h(n, m):
    return n*n/(m-1) + 2*n*m + n*np.power(2,m)
```

```python
x_f = np.array(range(5,2001))

y_2f = [h(x,2) for x in x_f]
y_3f = [h(x,3) for x in x_f]
y_4f = [h(x,4) for x in x_f]
y_5f = [h(x,5) for x in x_f]
y_6f = [h(x,6) for x in x_f]
y_7f = [h(x,7) for x in x_f]
y_8f = [h(x,8) for x in x_f]
```

```python
layout_1 = go.Layout(
    margin=dict(l=65, r=50, b=65, t=90),
    autosize=False,
    width=800,
    height=400,
    xaxis = dict(title='circuit wires n'),
    yaxis = dict(title='circuit size', range=[-80000,1200000]),
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
    )

trace2f = go.Scatter(x=x_f, y=y_2f, name= 'm=2', mode='lines', line=dict(color='#636EFA'))
trace3f = go.Scatter(x=x_f, y=y_3f, name= 'm=3', mode='lines', line=dict(color='#EF553B'))
trace4f = go.Scatter(x=x_f, y=y_4f, name= 'm=4', mode='lines', line=dict(color='#00CC96'))
trace5f = go.Scatter(x=x_f, y=y_5f, name= 'm=5', mode='lines', line=dict(color='#AB63FA'))
trace6f = go.Scatter(x=x_f, y=y_6f, name= 'm=6', mode='lines', line=dict(color='#FFA15A'))
trace7f = go.Scatter(x=x_f, y=y_7f, name= 'm=7', mode='lines', line=dict(color='#19D3F3'))
trace8f = go.Scatter(x=x_f, y=y_8f, name= 'm=8', mode='lines', line=dict(color='#FF6692'))

data = [trace2f, trace3f,trace4f,trace5f, trace6f, trace7f, trace8f]
fig = go.Figure(data = data, layout=layout_1)
fig.show()
```

```python
output_dir = r'resources\section_size_benchmark'
output_file_prefix = 'SectionSizeBenchmark'

output_file_timestamp_low = '170321_2142'
output_file_timestamp_mid = '170321_2148'
output_file_timestamp_high = '170321_2219'
output_file_timestamp_high_10 = '170321_2314'
output_file_timestamp_mid_1 = '170321_2328'
output_file_timestamp_high_1 = '170321_2350'
output_file_timestamp_miss_5 = '120421_2114'
```

```python
section_size_benchmark_low_df = pd.read_csv(rf'..\{output_dir}\{output_file_prefix}_{output_file_timestamp_low}.csv')
section_size_benchmark_mid_df = pd.read_csv(rf'..\{output_dir}\{output_file_prefix}_{output_file_timestamp_mid}.csv')
section_size_benchmark_high_df = pd.read_csv(rf'..\{output_dir}\{output_file_prefix}_{output_file_timestamp_high}.csv')
section_size_benchmark_high_10_df = pd.read_csv(rf'..\{output_dir}\{output_file_prefix}_{output_file_timestamp_high_10}.csv')
section_size_benchmark_mid_1_df = pd.read_csv(rf'..\{output_dir}\{output_file_prefix}_{output_file_timestamp_mid_1}.csv')
section_size_benchmark_high_1_df = pd.read_csv(rf'..\{output_dir}\{output_file_prefix}_{output_file_timestamp_high_1}.csv')
section_size_benchmark_miss_5_df = pd.read_csv(rf'..\{output_dir}\{output_file_prefix}_{output_file_timestamp_miss_5}.csv')

section_size_comparison_df = section_size_benchmark_low_df\
    .append(section_size_benchmark_mid_df)\
    .append(section_size_benchmark_high_df)\
    .append(section_size_benchmark_high_10_df)\
    .append(section_size_benchmark_mid_1_df)\
    .append(section_size_benchmark_high_1_df)\
    .append(section_size_benchmark_miss_5_df)

section_size_comparison_df = section_size_comparison_df.loc[:,['size', 'sec_size', 'num_gates']]
section_size_comparison_df.head()
```

```python
m_meas = np.array([3,3,4,4,5,5,6,6,7,7])
n_meas = np.array([12,58,59,217,218,681,682,1977,1978,2000])

n_vec_2d = np.array(range(2,2001))
m_2d_default = ([0.5]*1999) * np.log2(n_vec_2d)
```

```python
layout_3 = go.Layout(
    margin=dict(l=65, r=50, b=65, t=90),
    autosize=False,
    width=800,
    height=400,
    xaxis = dict(title='circuit wires n'),
    yaxis = dict(title='section width m', range=[1, 7.5]),
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    )

trace2 = go.Scatter(x= n_vec_2d, y = np.rint(m_2d_default), mode='lines', name='m = round(0.5 log_2(n)', line=dict(color='#EF553B'))
trace3 = go.Scatter(x= n_meas, y = m_meas, name='observed optimal m', mode='lines', line=dict(color='#00CC96'))

data = [trace3, trace2]
fig = go.Figure(data = data, layout=layout_3)
fig.show()
```

```python
section_meas_3_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 12) & (section_size_comparison_df['size'] <= 58) & (section_size_comparison_df['sec_size'] == 3)]
section_meas_4_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 59) & (section_size_comparison_df['size'] <= 217) & (section_size_comparison_df['sec_size'] == 4)]
section_meas_5_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 218) & (section_size_comparison_df['size'] <= 681) & (section_size_comparison_df['sec_size'] == 5)]
section_meas_6_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 682) & (section_size_comparison_df['size'] <= 1977) & (section_size_comparison_df['sec_size'] == 6)]
section_meas_7_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 1978) & (section_size_comparison_df['size'] <= 2000) & (section_size_comparison_df['sec_size'] == 7)]

section_meas_df = section_meas_3_df\
    .append(section_meas_4_df)\
    .append(section_meas_5_df)\
    .append(section_meas_6_df)\
    .append(section_meas_7_df)\
    .sort_values(by=['size'])

section_meas_df.head()
```

```python
section_pap_1_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 5) & (section_size_comparison_df['size'] <= 7) & (section_size_comparison_df['sec_size'] == 1)]
section_pap_2_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 8) & (section_size_comparison_df['size'] <= 32) & (section_size_comparison_df['sec_size'] == 2)]
section_pap_3_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 33) & (section_size_comparison_df['size'] <= 127) & (section_size_comparison_df['sec_size'] == 3)]
section_pap_4_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 128) & (section_size_comparison_df['size'] <= 512) & (section_size_comparison_df['sec_size'] == 4)]
section_pap_5_df = section_size_comparison_df[(section_size_comparison_df['size'] >= 513) & (section_size_comparison_df['size'] <= 2000) & (section_size_comparison_df['sec_size'] == 5)]

section_pap_df = section_pap_1_df\
    .append(section_pap_2_df)\
    .append(section_pap_3_df)\
    .append(section_pap_4_df)\
    .append(section_pap_5_df)\
    .sort_values(by=['size'])

section_pap_df.head()
```
