```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
output_dir = r'resources\runtime_performance'
output_file_prefix = 'RuntimePerformance'
output_file_timestamp = '210221_1943'
```

```python
df = pd.read_csv(rf'..\{output_dir}\{output_file_prefix}_{output_file_timestamp}.csv')

df_agg = df.groupby(['size', 'sec_size']).agg(num_gates_mean=('num_gates', 'mean'), num_gates_std=('num_gates', 'std'),
                                              process_time_mean=('process_time', 'mean'), process_time_std=('process_time', 'std'),
                                              initial_rss_mean=('initial_rss', 'mean'), initial_rss_std=('initial_rss', 'std'),
                                              final_rss_mean=('final_rss', 'mean'), final_rss_std=('final_rss', 'std'))

df_agg.head()
```

```python
sizes = [size for (size, _) in df_agg.index.tolist()]

sec_sizes = 500*np.array([sec_size for (_, sec_size) in df_agg.index.tolist()])

print(sec_sizes)
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Scatter(
        x=sizes,
        y=df_agg['num_gates_mean'].tolist(),
        error_y=dict(
            type='data', # value of error bar given in data coordinates
            array=df_agg['num_gates_std'].tolist(),
            visible=True)
    ))

fig.add_trace(go.Scatter(
        x=sizes,
        y=sec_sizes))

fig.show()
```
